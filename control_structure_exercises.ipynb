```python
#2C. break and continue

#i. Prompt the user for an odd number between 1 and 50. Use a loop and a break statement to continue 
#prompting the user if they enter invalid input. (Hint: use the isdigit method on strings to determine this). 
#Use a loop and the continue statement to output all the odd numbers between 1 and 50, except for the 
#number the user entered.

odd_number = input("Please choose an odd number between 1 and 50: ")
while True:
    if odd_number.isdigit() == False:
        print('[INVALID ENTRY: ENTRY NOT A NUMBER.]')
        odd_number = input("Please choose an odd a number between 1 and 50: ")
    elif int(odd_number) not in range(0,51):
        print('[INVALID ENTRY: NUMBER WAS NOT BETWEEN 1 AND 50.]')
        odd_number = input("Please choose an odd number between 1 and 50: ")
    elif int(odd_number) % 2 == 0:
        print('[INVALID ENTRY: NUMBER WAS NOT ODD.]')
        odd_number = input("Please choose an odd number between 1 and 50: ")

    else:
        break
print('\nNumber to skip is:', odd_number, '\n')
for n in range(51):
    if n == int(odd_number): ##skip designated number
        continue
    elif n % 2 == 0: ##skip evens
        continue
    else:
        print('Here is an odd number:', n)
```

```python
#2D. The input function can be used to prompt for input and use that input in your python code. Prompt the user to enter a positive number and write a loop that counts from 0 to that number. (Hints: first make sure that the value the user entered is a valid number, also note that the input function returns a string, so you'll need to convert this to a numeric type.)

positive_number= input('Please enter a positive number: ')
while True:
    if positive_number.isdigit() == False:
        print('[INVALID ENTRY: ENTRY NOT A NUMBER.]')
        positive_number= input('Please enter a positive number: ')
    elif int(positive_number) < 0:
        print('[INVALID ENTRY: NUMBER WAS NOT A POSITIVE NUMBER.]')
        positive_number= input('Please enter a positive number: ')
    else:
        break
print('\nNumber to count to is:', positive_number, '\n')
for n in range(51):
    if n > int(positive_number):
        break
    print(n)
```

```python
#2E. Write a program that prompts the user for a positive integer. Next write a loop that prints out the numbers from the number the user entered down to 1.
positive_num= input('Please enter a positive number: ')
while True:
    if positive_num.isdigit() == False: #check if it is a number
        print('[INVALID ENTRY: ENTRY NOT A NUMBER.]')
        positive_num= input('Please enter a positive number: ')
    elif int(positive_num) <= 1:  #check if it is positive
        print('[INVALID ENTRY: NUMBER WAS NOT A POSITIVE NUMBER.]')
        positive_num= input('Please enter a positive number: ')
    else:
        break
print('\nNumber to count down from is:', positive_num, '\n')
p_num = int(positive_num)
while p_num>= 1: 
    print(p_num)
    p_num -= 1
```

```python
#3. Fizzbuzz

#One of the most common interview questions for entry-level programmers is the FizzBuzz test. Developed by Imran Ghory, the test is designed to test basic looping and conditional logic skills.
    #Write a program that prints the numbers from 1 to 100.
    #For multiples of three print "Fizz" instead of the number
    #For the multiples of five print "Buzz".
    #For numbers which are multiples of both three and five print "FizzBuzz".
n=0
for n in range(1,101):
    if (n%3) + (n%5) == 0: #if divisible by 3 AND 5 print FIZBUZZ instead
        print('FizzBuzz')
    elif n % 3 == 0: #if divisible by 3 print FIZZ instead
        print('Fizz')
    elif n % 5 == 0: #if divisible by 5 print BUZZ instead
        print ('Buzz')
    else:
        print(n)
```

```python
#5. Convert given number grades into letter grades.

    #Prompt the user for a numerical grade from 0 to 100.
    #Display the corresponding letter grade.
    #Prompt the user to continue.
    #Assume that the user will enter valid integers for the grades.
    #The application should only continue if the user agrees to.
    #Grade Ranges:
        #A : 100 - 88
        #B : 87 - 80
        #C : 79 - 67
        #D : 66 - 60
        #F : 59 - 0


lemniscate = True

while lemniscate == True:
    user_grade = input('Please insert the numerical grade you recieved: ') 
    if int(user_grade) in range(88,101):
        print('You got an A!')
    elif int(user_grade) in range(80,88):
        print('You got an B!')
    elif int(user_grade) in range(67,80):
        print('You got an C.')
    elif int(user_grade) in range(60,67):
        print('You got an D.')
    else:
        print("You got an F.")
    quit_query= input("Would you like to continue putting in grades? Yes = 1, No = 0 ")
    lemniscate = int(quit_query)
    
```

```python
#6. Create a list of dictionaries where each dictionary represents a book that you have read. Each dictionary in the list should have the keys title, author, and genre. Loop through the list and print out information about each book.

my_favorite_books= [
{'Title':'Warrior', 'Author':'Marie Brennan', 'Genre':'Fantasy'} ,
{'Title':'Witch', 'Author':'Marie Brennan', 'Genre':'Fantasy'} ,
{'Title':'Talking to Dragons', 'Author':'Patricia C. Wrede', 'Genre':'Fantasy'} ,
{'Title':'Halo: The Fall of Reach', 'Author':'Eric Nylund', 'Genre':'Science Fiction'} ,
{'Title':'Ready Player One', 'Author':'Ernest Cline', 'Genre':'Science Fiction'}
]


for x in my_favorite_books: #where x represents an element in the list, a single dictionary, or for this example, a single book
    for key in x: #where key represents the literal key in the current dictionary(or book) we are currently looking at
        print(key,':', x[key])
    print('\n')
    
```

```python
#6A. Prompt the user to enter a genre, then loop through your books list and print out the titles of all the books in that genre.

user_genre= input('What genre of book are you interested in? ')
found_flag = False #if we find a matching genre, this will become true, if not, then the 'for' loop will iterate without printing anything and then the 'if' loop will activate and let the user know nothing was found

for book in my_favorite_books:
    if book['Genre'] == user_genre:
        found_flag = True 
        for key in book:
            print(key + ': ' + book[key])
        print()
            
if found_flag == False: 
    print('There are no suggestions for this genre!')
    
          
```
